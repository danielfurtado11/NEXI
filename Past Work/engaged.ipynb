```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Supondo que os dados foram lidos de um arquivo CSV chamado 'data.csv'
data = pd.read_csv('2weeks_new.csv')

# Ordenar os dados por pessoa registada e por tempo (antigo para recente)
data = data.sort_values(by=['person', '_time'])

# Converter a coluna de tempo para datetime com o formato misto
data['_time'] = pd.to_datetime(data['_time'], format='mixed')
```

```python
# Adicionar o dia de cada registo (já que existem)

# CORIRGIR para conjunto ano-mes-dia
data['small_date'] = data['_time'].apply(lambda x: x.strftime('%Y-%m-%d'))
data['hour'] = data['_time'].apply(lambda x: x.hour + 1)
data['number_day_week'] = data['_time'].apply(lambda x: x.weekday())
data['day_week'] = data['_time'].apply(lambda x: x.day_name())
data['week_year_number'] = data['_time'].apply(lambda x: x.isocalendar()[0] * 100 + x.isocalendar()[1])
data['weekend'] = data['day_week'].apply(lambda x: x in ['Saturday', 'Sunday'])
```

```python
# Criar ficheiro EXCEL para guardar estes resultados obtidos

data.to_excel('2weeks_complete.xlsx', index=False)
#data.to_csv('engagement_percentages.csv', index=False)
```

```python
# Função para calcular a percentagem de tempo de engagement
def calculate_engagement_percentage(group):
    total_time = (group['_time'].max() - group['_time'].min()).total_seconds()

    if total_time == 0:
        return 0

    engaged_time = 0

    previous_time = group['_time'].iloc[0]
    previous_engaged = group['engaged'].iloc[0]

    for current_time, current_engaged in zip(group['_time'][1:], group['engaged'][1:]):
        interval = (current_time - previous_time).total_seconds()
        #print(interval)
        if previous_engaged == 'yes':
            engaged_time += interval if current_engaged == 'yes' else interval / 2
        elif current_engaged == 'yes':
            engaged_time += interval / 2
        previous_time = current_time
        previous_engaged = current_engaged

    # Considerar o último intervalo
    if previous_engaged == 'yes':
        engaged_time += (group['_time'].iloc[-1] - previous_time).total_seconds() / 2

    return (engaged_time / total_time)

# Aplicar a função para cada grupo de pessoa
engagement_percentages = data.groupby(['person','small_date','day_week','number_day_week','weekend','week_year_number','hour','gender']).apply(calculate_engagement_percentage)

# Exibir os resultados
print(engagement_percentages)
```

```python
# Calcular o número de pessoas por hora e dia de registo (para depois ser mais fácil calcular a média, mínimo e máximos de cada dia)

person_per_hour = engagement_percentages.groupby(['small_date', 'hour']).size().reset_index(name='count_person_hour')
person_per_hour.to_excel('2weeks_count_person_hour.xlsx', index=False)
```

```python
# Função para calcular a percentagem de tempo de engagement
def calculate_engagement_times(group):
    total_time = (group['_time'].max() - group['_time'].min()).total_seconds()

    if total_time == 0:
        return 0.0

    engaged_time = 0.0

    previous_time = group['_time'].iloc[0]
    previous_engaged = group['engaged'].iloc[0]

    for current_time, current_engaged in zip(group['_time'][1:], group['engaged'][1:]):
        interval = (current_time - previous_time).total_seconds()
        #print(interval)
        if previous_engaged == 'yes':
            engaged_time += interval if current_engaged == 'yes' else interval / 2
        elif current_engaged == 'yes':
            engaged_time += interval / 2
        previous_time = current_time
        previous_engaged = current_engaged

    # Considerar o último intervalo
    if previous_engaged == 'yes':
        engaged_time += (group['_time'].iloc[-1] - previous_time).total_seconds() / 2

    return engaged_time

# Aplicar a função para cada grupo de pessoa
engagement_times = data.groupby(['person','small_date','day_week','number_day_week','weekend','week_year_number','hour','gender']).apply(calculate_engagement_times)

# Exibir os resultados
print(max(engagement_times))
```

```python
# Verificar se houve mudança na coluna "engaged" de 'no' para 'yes' para cada pessoa, para cada dia

# Função para verificar se houve mudança de 'no' para 'yes' na coluna 'engaged'
def check_engagement_change(group):
    #return any((engaged_series.shift() == 'no') & (engaged_series == 'yes'))
    engaged_series = group['engaged']

    engaged_series = engaged_series.reset_index(drop=True)
    
    # Check if the series is not empty
    # len(engaged_series) == 0
    if all(engaged_series == 'no'):
        return 'Never engaged - No effect'
    
    elif all(engaged_series == 'yes'):
        return 'Always engaged'

    elif all(engaged_series[0:2] == 'no') and any(engaged_series[3:] == 'yes'):
        return 'Positive effect (no to yes)'
    
    else:
        return 'Engaged, but Left (yes to no)'

# Aplicar a função para cada grupo de pessoa
engagement_changes = data.groupby(['person','small_date','day_week','number_day_week','weekend','week_year_number','hour','gender']).apply(check_engagement_change)

# Exibir os resultados
print(engagement_changes)
```

```python
# Criar ficheiro EXCEL para guardar estes resultados obtidos

file_engagement_changes = engagement_changes.reset_index(name='engaged_change')
file_engagement_changes.to_excel('engagement_changes.xlsx', index=False)
#engagement_changes.to_csv('engagement_changes.csv', index=False)
```

```python
# Construir gráfico circular da lisat anterior

# Contar os valores True e False
contagem = engagement_changes.value_counts()
labels = contagem.index#.map(lambda x: 'True' if x else 'False')
valores = contagem.values

fig, ax = plt.subplots()
ax.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('A pessoa passsou a ficar interessada (engaged)?')
plt.show()
```
